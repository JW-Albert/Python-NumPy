# N-D Array Shape

```python
import numpy as np
```

## Basic

```python
data_1 = np.ones(10)
print("data_1 shape =", data_1.shape)
```

```python
data_2 = np.ones([2 ,3])
print("data_2 shape =" ,data_2.shape)
```

## Rotate

```python
data_3 = np.array([
    [1, 2, 3],
    [4 ,5 ,6]
])
print("data_3 data:")
print(data_3)

print("data_3 shape =" ,data_3.shape ,"\n")


print("Rotated data_3 data:")
print(data_3.T)
print("Rotated data_3 shape =",data_3.T.shape)
```

## Reduce Degree - ravel()

```python
data_4 = np.array([
    [
        [1 ,2] ,[2 ,3]
    ],
    [
        [3 ,4] ,[4 ,5]
    ]
])
print("data_4 data:")
print(data_4)
print("data_4 shape =" ,data_4.shape ,"\n")
data_5 = data_4.ravel()
print("raveled data_4:")
print(data_5)
```

## 重塑資料形狀

```python
data_4 = np.array([
    [
        [1 ,2] ,[2 ,3]
    ],
    [
        [3 ,4] ,[4 ,5]
    ]
]) # 2 * 2 * 2
print("data_4 data:")
print(data_4)
print("data_4 shape =" ,data_4.shape ,"\n")

data_6 = data_4.reshape(2 ,4) # 2 * 4 = 2 * 2 * 2
print("Reshaped data_4:")
print(data_6)
print("data_6 shape =" ,data_6.shape ,"\n")
```

```python
data_7 = np.arange(1 ,11 ,1).reshape(2, 5)
print("data_7:\n", data_7)
```
