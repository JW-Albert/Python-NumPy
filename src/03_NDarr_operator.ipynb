# N-D Array Operator

```python
import numpy as np
```

## 逐元計算 (elementwise)
- 前提：陣列維度需要相等

```python
data_1 = np.array([1 ,2 ,3])
data_2 = np.array([4 ,5 ,6])
print("data_1 =", data_1)
print("data_2 =" ,data_2)
result = None
```

### add

```python
result = data_1 + data_2
print("data_1 + data_2 =" ,result)
```

### times

```python
result = data_1 * data_2
print("data_1 * data_2 =", result)
```

### bigger

```python
result = data_1 > data_2
print("data_1 > data_2 =", result)
```

### equal

```python
result + data_1 == data_2
print("data_1 == data_2 =", result)
```

## 矩陣計算 (matrix)

```python
data_3 = np.array([
    [1 ,2]
]) # 1*2
data_4 = np.array([
    [1, 2, 3],
    [2, 3, 4]
]) # 2*3
print("data_3 =" ,data_3)
print("data_4 =" ,data_4)
result = None
```

### dot

```python
result = data_3.dot(data_4)
print("data_3 dot data_4 =", result)
result = data_3@data_4
print("data_3 dot data_4 =", result)
```

### cross

```python
result = np.outer(data_3 ,data_4)
print("data_3 cross data_4 =", result)
```
