# N-Dimensional Array(NDArray)

```python
import numpy as np
```

## 1-D array

### Directly from list

```python
oneD_arr = np.array([1 ,2 ,3, 4])
print("1D array data =", oneD_arr)
```

### empty array

```python
empty_1D = np.empty(4)
print("Empty 1D array =", empty_1D)
```

### zeros array

```python
zeros_1D = np.zeros(4)
print("Zeros 1D array =", zeros_1D)
```

### ones array

```python
ones_1D = np.ones(4)
print("Ones 1D array =", ones_1D)
```

### arange array

```python
arange_1D = np.arange(0 ,5 ,1) # or arange(5)
print("Arrange 1D array =", arange_1D)
```

## 2-D array

```python
twoD_arr = np.array([
    [0 ,1 ,2],
    [1 ,2 ,3],
    [2 ,3 ,4]
])
print("2D array data =" ,twoD_arr)
```

### empty array

```python
empty_2D = np.empty([2 ,2]) # 2*2 empty array
print("Empty 2D array =", empty_2D)
```
